Now for detailed analysis I am geting a Age and other data for indepth analysis of the Customer and order details

```python
import pandas as pd
from datetime import datetime

# Calculate age using only the year
Final_Electronic_data['Age'] = datetime.now().year - Final_Electronic_data['Birthday'].dt.year

# Extract day name
Final_Electronic_data['Purchase Day Name'] = Final_Electronic_data['Order Date'].dt.day_name()

# Extract month name
Final_Electronic_data['Purchase Month Name'] = Final_Electronic_data['Order Date'].dt.month_name()

Final_Electronic_data.head()
```
